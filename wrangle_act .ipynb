# Assess

## Quality

* Lots of NaN values in various columns
* rating numerator and denominator are ints, should be floats (some observations have decimals)
* "Timestamp" column is wrong datatype, should be datetime, is string
* "source" column has unnecessary HTML; anchor tag with the source inside: Twitter mobile, vine, Twitter Web or Tweetdeck
* "retweeted_status_timestamp" column is wrong datatype, should be datetime, is string
* "tweet_id" should be a string, not an int
* "in_reply_to_user_id" is wrong datatype, should be string, not float
* "in_reply_to_status_id" is wrong datatype, should be string, not float
* "img_num" column in images dataframe is meaningless, all rows have one image regardless of the number in img_num
* Retweets need to be removed so we don't have duplicated data

## Tidiness
* "doggo, floofer, pupper and puppo" columns are redundant. Could be reworked into one column "dog_type"
* Image predictions dataframe column names are not descriptive
* 3 separate dataframes that could be inner joined

# Clean

```python
# Not much we can do about the NaN values for retweets and replies and such, as this means the tweet 
# was not a reply or a retweet.

# Let's make copies of the unclean datasets for cleaning.

df_clean = df.copy()
images_clean = images.copy()
df_json_clean = df_json.copy()


```

## Define

* Retweets need to be removed so we don't have duplicated data

## Code

```python
# remove retweets from dataframes

df_clean = df_clean[pd.isnull(df_clean.retweeted_status_id)]
df_clean = df_clean[pd.isnull(df_clean.retweeted_status_timestamp)]
df_clean = df_clean[pd.isnull(df_clean.retweeted_status_user_id)]
```

## Test

```python
df_clean.info()
```

## Define

* "doggo, floofer, pupper and puppo" columns are redundant. Could be reworked into one column "dog_type"

## Code

```python
# create new column for dog keywords
df_clean['dog_keyword'] = df_clean['doggo'].replace('None', '') + df_clean['puppo'].replace('None','') + df_clean['pupper'].replace('None', '') + df_clean['floofer'].replace('None','')
```

```python
#drop unnecessary columns
df_clean.drop(['doggo','floofer','pupper','puppo'], axis=1, inplace=True)
```

## Test

```python
df_clean['dog_keyword'].value_counts()
```

## Define

* rating numerators are ints, should be floats (some observations have decimals)

## Code

```python
# change ratings to floats

import re

df_clean['rating_numerator'] = df_clean['text'].str.extract(r'(\d+(\.\d+)?)\/\d+').astype(float)

```

## Test

```python
df_clean.info()
```

## Define
* "Timestamp" column is wrong datatype, should be datetime, is string

## Code

```python
# Import datetime module to fix data types for timestamps
from datetime import datetime

df_clean['timestamp'] = pd.to_datetime(df_clean['timestamp'])
```

## Test

```python
type(df_clean['timestamp'][0])
```

## Define
* "retweeted_status_timestamp" column is wrong datatype, should be datetime, is string

## Code

```python
# Fix wrong datatypes for retweeted status timestamp column
df_clean['retweeted_status_timestamp'] = pd.to_datetime(df_clean['retweeted_status_timestamp'])
```

## Test

```python
df_clean.info()
```

## Define
* "source" column has unnecessary HTML; anchor tag with the source inside: Twitter mobile, vine, Twitter Web or Tweetdeck

## Code

```python
# use regular expression to scrape away the unwanted html tag for the source column

regex_pat = re.compile(r'<[^>]*>') 
df_clean['source'] = df_clean['source'].str.replace(regex_pat, '')
```

## Test

```python
df_clean['source'].value_counts()
```

## Define
* "tweet_id" should be a string, not an int

## Code

```python
# convert tweet id to string

df_clean['tweet_id'] = df_clean['tweet_id'].astype(str)
images_clean['tweet_id'] = images_clean['tweet_id'].astype(str)
df_json_clean['tweet_id'] = df_json_clean['tweet_id'].astype(str)
```

## Define
* "in_reply_to_user_id" is wrong datatype, should be string, not float

## Code

```python
# convert reply to user id to string

df_clean['in_reply_to_user_id'] = df_clean['in_reply_to_user_id'].astype(str)
```

## Test

```python
df_clean.info()
```

## Define
* "in_reply_to_status_id" is wrong datatype, should be string, not float

## Code

```python
#convert reply to status id to string

df_clean['in_reply_to_status_id'] = df_clean['in_reply_to_status_id'].astype(str)
```

## Test

```python
df_clean.info()
```

## Define
* img_num column in the images dataframe is useless

## Code

```python
# remove useless img_num column

images_clean.drop('img_num', axis = 1, inplace = True)
```

## Test

```python
images_clean.head()
```

## Define
* images column names are not descriptive

## Code

```python
# rename columns in images_clean to be more descriptive

images_clean.rename(columns={'p1':'prediction_1', 'p1_conf':'prediction_1_confidence', 'p1_dog':'prediction_1_dog', 'p2':'prediction_2', 'p2_conf':'prediction_2_confidence', 'p2_dog':'prediction_2_dog', 'p3':'prediction_3', 'p3_conf':'prediction_3_confidence', 'p3_dog':'prediction_3_dog'}, inplace=True)
```

* Three separate dataframes that could be inner joined

## Code

```python
df_final = pd.merge(df_clean, images_clean, on='tweet_id', how='inner')
df_final = pd.merge(df_final, df_json_clean, on='tweet_id', how='inner')
```

# Visualize

```python
# import matplotlib for visualizations
import matplotlib.pyplot as plt


```

```python
#create a pie chart for the keywords of dogs

%matplotlib inline

plt.pie(df_clean['dog_keyword'].value_counts(), autopct='%1.1f%%', labels = ['None', 'pupper', 'doggo', 'puppo', 'doggopupper', 'floofer', 'doggopuppo', 'doggofloofer']);
plt.title('Dog Keywords');
```

We see from the pie chart that most dogs don't have keywords, and of the ones that do, most are "pupper" and "doggo", with less being "floofer" and "puppo", and even less having multiple keywords.

```python
#create a histogram for distribution of favorite counts

plt.hist(df_json['favorite_count'], bins=10, edgecolor='black');
plt.xlabel('Favorite Count')
plt.ylabel('Frequency')
plt.title('Number of Favorite Counts for WeRateDogs tweets')
plt.show();
```

This histogram shows us that most tweets have between ~10000 and 30000 favorites, with very few reaching up higher towards even 140k.

```python
#create a histogram for distribution of retweet counts

plt.hist(df_json['retweet_count'], bins=10, edgecolor='black');
plt.xlabel('Retweet Count')
plt.ylabel('Frequency')
plt.title('Number of Retweet Counts for WeRateDogs tweets')
plt.show();
```

This histogram is very similar to the distribution of the favorites one, with most tweets having between 1000-10000 retweets, and few going as high as 50k.

```python
# create a histogram for distribution of rating numerators

plt.hist(df_clean.query('rating_numerator < 20')['rating_numerator'], edgecolor='black')
plt.xlabel('Numerator')
plt.ylabel('Frequency')
plt.title('Histogram of Numerator Values')
plt.show();
```

From this histogram, we can see that most rating numerators fall within the 11-13 range.

```python
plt.hist(df_clean['source'], edgecolor='black')
plt.xlabel('Source')
plt.ylabel('Frequency')
plt.title('Distribution of Sources for Accessing WeRateDogs Tweets')
plt.show();
```

From this histogram, we see that an overwhelming majority of WeRateDogs twitter views come from iPhones, rather than the other options.

```python
plt.pie(images_clean['prediction_1_dog'].value_counts(), labels=['True', 'False'], autopct = '%1.1f%%', explode =(0, 0.1))
plt.title('True vs False of machine learning predictions for dog images');
```

This pie chart shows us that the machine learning algorithm that made the predictions for the WeRateDogs tweets of the dog's breeds is accurate roughly 74% of the time.
